```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url="https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/customer_segmentation.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.drop("Address", axis = 1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = df.values[:, 1:]
X = np.nan_to_num(X)
data = StandardScaler().fit_transform(X)
data
```

```python
k = 3
k_means = KMeans(n_clusters=k)
k_means.fit(data)
print(k_means.labels_)
```

```python
df['cluster'] = k_means.labels_
df.head()
```

```python
df['cluster'].value_counts()
```

```python
df.groupby('cluster').mean()
```

```python
area = np.pi * (X[:, 1])**2
plt.scatter(X[:, 0], X[:, 3], s = area, c = k_means.labels_.astype(float), alpha = 0.5)
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize = (10, 10))
plt.clf()

ax = fig.add_subplot(111, projection='3d')

plt.cla()
# plt.ylabel('Age', fontsize=18)
# plt.xlabel('Income', fontsize=16)
# plt.zlabel('Education', fontsize=16)
ax.set_xlabel('Years Employed')
ax.set_ylabel('Age')
ax.set_zlabel('Income')

ax.scatter(X[:, 2], X[:, 0], X[:, 3], c= k_means.labels_.astype(float))
```
