```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head(20)
```

```python
plt.figure(figsize = (10, 10))

sns.scatterplot(data = df, x = "TV", y = 'Sales' )

plt.show()

```

```python
sns.scatterplot(data = df, x = "Radio", y = 'Sales' )
```

```python

sns.scatterplot(data = df, x = "Newspaper", y = 'Sales' )
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size = 0.10, random_state=42)
```

```python
from sklearn import linear_model

lr_model = linear_model.LinearRegression()
```

```python
train_x = np.asarray(train_set[['TV']])
train_y = np.asarray(train_set[['Sales']])


lr_model.fit(train_x, train_y)
```

```python
theta1 = lr_model.coef_[0][0]
theta0 = lr_model.intercept_[0]
```

```python
print(theta1, theta0)
```

```python
plt.figure(figsize = (10, 8))
sns.scatterplot(data = df, y = 'Sales', x = 'TV')
plt.plot(train_x, theta0+theta1*train_x, '-r')
plt.show()
```

```python
test_x = np.asarray(test_set[['TV']])
tetx_y = np.asarray(test_set[['Sales']])
```

```python
y_predicted = lr_model.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_predicted, tetx_y)
```
