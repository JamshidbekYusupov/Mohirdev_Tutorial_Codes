```python
# importing necessary libraries
from fastai.vision.all import *
from ipywidgets import widgets
```

```python
# data
path = untar_data(URLs.PETS)/'images'
```

```python
# labeling function
def is_cat(x): return x[0].isupper()

#data loader
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), label_func=is_cat, item_tfms = Resize(224),
    valid = 0.2, seed = 42,
)

# training

learn = cnn_learner(dls, resnet34, metrics = accuracy)
learn.fine_tune(2)
```

With the trainded model

```python
#image uploading

upload = widgets.FileUpload()
upload
```

```python
img = PILImage.create(upload.data[-1])
#predicting the uploaded image
pred, _, prob = learn.predict(img)
print(f"Yuklangan rasm mushumki:{pred}")
print(f"Aniqlilik darajasi:{prob[1].item():.3f}")
img
```
